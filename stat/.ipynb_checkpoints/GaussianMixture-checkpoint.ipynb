```python
import numpy as np
from numpy.random import randint
from matplotlib import pyplot as plt
from scipy.stats import multivariate_normal as mn
import matplotlib
from numpy.linalg import inv
%matplotlib inline
```

```python
I = np.array([[1, 0], [0, 1]])
blue_mean_rv = mn([1, 0], I)
red_mean_rv = mn([0, 1], I)
blue_rv = [mn(blue_mean_rv.rvs(), I/5) for _ in range(10)]
red_rv = [mn(red_mean_rv.rvs(), I/5) for _ in range(10)]
rvs = {'blue': blue_rv, 'red': red_rv}
```

```python
def bayes(x):
    pr, pb = 0, 0
    for r in rvs['blue']:
        pb += r.pdf(x)
    for r in rvs['red']:
        pr += r.pdf(x)
    return pr/(pb+pr)
```

```python
def bayes_classifier(x):
    return bayes(x) > 0.5
```

```python
X, Y = np.zeros(shape=(10200, 2)), np.zeros(shape=(10200, 1))
for i in range(10200):
    X[i] = rvs[('blue','red')[i%2]][randint(10)].rvs()
    Y[i] = i%2
X_train, X_test = X[:200,:], X[200:,:]
Y_train, Y_test = Y[:200,:], Y[200:,:]
```

```python
x_min = np.amin(X, axis=0)
x_max = np.amax(X, axis=0)
```

```python
x1_grid, x2_grid = np.meshgrid(np.linspace(x_min[0], x_max[0], 100),
                               np.linspace(x_min[1], x_max[1], 100))
y_grid = np.zeros(shape=x1_grid.shape)
for i in range(100):
    for j in range(100):
        y_grid[i, j] = bayes_classifier([x1_grid[i,j], x2_grid[i,j]])
```

```python
fig, ax1 = plt.subplots(figsize=(6, 6), dpi=110)
cmap = matplotlib.colors.ListedColormap(['#4A81BF', '#C3514E'])
ax1.contour(x1_grid, x2_grid, y_grid, colors='black', linewidths=1)
ax1.scatter(x1_grid, x2_grid, c=y_grid, s=0.1, cmap=cmap)
ax1.scatter(X_train[:,0], X_train[:,1], c=Y_train[:,0], cmap=cmap)
ax1.set_xlim(x_min[0], x_max[0])
ax1.set_ylim(x_min[1], x_max[1])
```

```python
Y_test_hat_b = np.zeros(shape=(10000, 1))
for i in range(10000):
    Y_test_hat_b[i] = bayes(X_test[i])
```

```python
(Y_test - Y_test_hat_b).T.dot(Y_test - Y_test_hat_b) / 10000
```

```python
X, Y = np.ones(shape=(200, 3)), Y_train
X[:,1:] = X_train
```

```python
b = inv((X.T.dot(X))).dot(X.T).dot(Y).squeeze()
def linear(x):
    return b.dot(x)
def linear_classifier(x):
    return b.dot(x) > 0.5
```

```python
for i in range(100):
    for j in range(100):
        y_grid[i, j] = linear_classifier([1, x1_grid[i,j], x2_grid[i,j]])
```

```python
y1 = (0.5 - b[0] - x_min[0]*b[1])/b[2]
y2 = (0.5 - b[0] - x_max[0]*b[1])/b[2]
```

```python
fig, ax1 = plt.subplots(figsize=(6, 6), dpi=110)
ax1.plot([x_min[0], x_max[0]], [y1, y2], color='black')
ax1.scatter(x1_grid, x2_grid, c=y_grid, s=0.1, cmap=cmap)
ax1.scatter(X_train[:,0], X_train[:,1], c=Y_train[:,0], cmap=cmap)
ax1.set_xlim(x_min[0], x_max[0])
ax1.set_ylim(x_min[1], x_max[1])
```

```python
Y_test_hat_l = np.zeros(shape=(10000, 1))
for i in range(10000):
    Y_test_hat_l[i] = linear([1, X_test[i, 0], X_test[i, 1]])
```

```python
(Y_test - Y_test_hat_l).T.dot(Y_test - Y_test_hat_l) / 10000
```
